```python
def printThis(position):
    print("Leaving here")
    if position==0:
        print("Base condition got hit")
        return
    if position%2==0:
        print("Even position",position)
    else:
        print("Odd position",position)
    printThis(position-1)
    for index in range(position,0,-1):
        print("Index is",index)
    print("Entering here",position)
printThis(11)
```

```python
def find_max(lst):
    if len(lst) == 1:
        return lst[0]
    else:
        return max(lst[0], find_max(lst[1:]))

def find_min(lst):
    if len(lst) == 1:
        return lst[0]
    else:
        return min(lst[0], find_min(lst[1:]))

my_list = [5, 3, 9, 1, 7, 2]
print("Maximum element:", find_max(my_list))
print("Minimum element:", find_min(my_list))
```
